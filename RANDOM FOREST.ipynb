## RANDOM FOREST:
- A random forest is a data construct applied to machine learning that develops large numbers of random decision trees analyzing sets of variables. This type of algorithm helps to enhance the ways that technologies analyze complex data.
## Process:
- Samples seen by each deicision tree is not exactly same
- No. of columns for each decision tree is equal sqrt ( no. o original columns ). Columns will be randomly selected.
- Grow full or depth to have a max_depth
- For Each test sample , predict value from each decision tree

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
hr = pd.read_csv('HR Analytics.csv')
hr.head()
```

```python
# to convert categorical to numeric we use dummies
hr_dummies=pd.get_dummies(hr)
train , test = train_test_split(hr_dummies, test_size=0.3, random_state=100)

hr_dummies.head()
```

```python
train_y=train['Attrition']
test_y = test['Attrition']
train_x = train.drop('Attrition',axis=1)
test_x = test.drop('Attrition',axis=1)
```

```python
rf_model = RandomForestClassifier(random_state=100)
rf_model.fit(train_x,train_y)
test_pred = rf_model.predict(test_x)
```

```python
df_pred = pd.DataFrame({'actual':test_y,'predicted':test_pred})
df_pred['pred_status']  = df_pred['actual'] ==df_pred['predicted']
print('MODEL ACCURACY: ',+df_pred['pred_status'].sum()/df_pred.shape[0]*100)
```

```python
tp2 = df_pred[(df_pred['predicted']==1) &(df_pred['actual']==1)].shape[0]

# TRUE NEGATIVE

tn2 = df_pred[(df_pred['predicted']==0) &(df_pred['actual']==0)].shape[0]

# FALSE POSITIVE

fp2 = df_pred[(df_pred['predicted']==1) &(df_pred['actual']==0)].shape[0]

#FALSE NEGATIVE

fn2 = df_pred[(df_pred['predicted']==0) &(df_pred['actual']==1)].shape[0]


print('TRUE POSITIVE: %d' %tp2)
print('TRUE NEGATIVE: %d' %tn2)
print('FALSE POSITIVE: %d' %fp2)
print('FALSE NEGATIVE: %d' %fn2)
```

```python

Accuracy= (tp2+tn2)/(tp2+tn2+fp2+fn2)
Sensitivity=tp2/(tp2+fn2)
Specificity=tn2/(fp2+tn2)
```

```python
rf=pd.Series({'TP':tp2,'TN':tn2,'FP':fp2, 'FN':fn2, 'Accuracy': Accuracy, 'Sensitivity':Sensitivity, 'Specificity':Specificity})
```

```python
RandomForest=pd.DataFrame({'RandomForest':rf})
RandomForest
```
