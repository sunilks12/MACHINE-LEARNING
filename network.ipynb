```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt 
```

```python
def visualize_network(file_path):
    import pandas as pd
    import numpy as np
    import networkx as nx
    import matplotlib.pyplot as plt
    
    df = pd.read_csv(file_path)
    df = df[['user_username','Value','interaction_type']]
    df['nodes'] = df[['user_username', 'Value']].apply(tuple, axis=1)
    
    g = nx.DiGraph()  
    g.add_edges_from(df['nodes'])
    degrees = dict(nx.degree(g))
    node_sizes = [v * 10 for v in degrees.values()]
    plt.figure(figsize = (20, 10))
    nx.draw(g, with_labels = False, node_size =node_sizes, font_size = 12)  
    plt.show()  
    return None

filepath='final_data.csv'
visualize_network(filepath)
```

```python
def metric_calculation():    
    degree = nx.degree_centrality(g)
    between = nx.betweenness_centrality(g)
    close = nx.closeness_centrality(g)
    out_degree = nx.out_degree_centrality(g)
    in_degree = nx.in_degree_centrality(g)
    df_final = pd.DataFrame([degree, between,close,in_degree,out_degree]).T
    df_final.reset_index(inplace = True)
    df_final.columns = ['user_username','degree','between','close','in_degree','out_degree']
    df_final['user_name']=df_final['user_username']
    df_final=df_final.drop('user_username',axis=1)
    first_column = df_final.pop('user_name')
    df_final.insert(0, 'user_name', first_column)
    return df_final
result = metric_calculation()
result.head(3)
```

```python
library('readr')


sow.data=read_csv("CreditCard_SOW_data.csv")
sow.data$logIncome= log(sow.data$Income)

sow.data$logSowRatio= log(sow.data$WalletShare/(1-sow.data$WalletShare))
sow.data$ConsumerID = as.factor(sow.data$ConsumerID)


sow.data.lm1 =  lm(logSowRatio~History+Balance+Promotion+History*Promotion+logIncome*Promotion, 
             data=sow.data)

summary(sow.data.lm1) 

AIC(sow.data.lm1)


library('lme4')

sow.data.lm2 <- lmer(sow.data$logSowRatio ~ sow.data$History+sow.data$Balance + sow.data$Promotion + 
               sow.data$History:sow.data$Promotion+sow.data$logIncome:sow.data$Promotion +
                 (1 + sow.data$Promotion | sow.data$ConsumerID),
                data=sow.data,control=lmerControl(optimizer="bobyqa"))

summary(sow.data.lm2)

AIC(sow.data.lm2)




'''
In this case, the difference in AIC between model 1 and model 2 is -1085 - (-2656) = 1571. This indicates a large effect and 
suggests that model 2 is a significantly better fit for the data than model 1.
'''

library("MCMCpack")


sow.ba1 = 
  MCMCregress(logSowRatio~History+Income+Balance+Promotion, 
              mcmc=6000, thin=6, data=sow.data)
summary(sow.ba1)





sow.ba2 = MCMChregress(fixed=logSowRatio~History+Balance+Promotion+History:Promotion+logIncome:Promotion,
                random=~Promotion, group="ConsumerID", mcmc=6000, 
               thin=6, data=sow.data, r=3, R=diag(2))

summary(sow.ba2$mcmc[,1:6])



```
